```python

!pip install tensorflow matplotlib

import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Load images
content_path = "drawing.jpg"
style_path = "color.jpg"


def load_img(path):
    max_dim = 512
    img = tf.io.decode_image(tf.io.read_file(path), channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    shape = tf.cast(tf.shape(img)[:-1], tf.float32)
    long_dim = max(shape)
    scale = max_dim / long_dim
    new_shape = tf.cast(shape * scale, tf.int32)

    img = tf.image.resize(img, new_shape)
    img = img[tf.newaxis, :]
    return img

content_image = load_img(content_path)
style_image = load_img(style_path)

# Display images
plt.subplot(1,2,1)
plt.imshow(tf.squeeze(content_image))
plt.title("Content Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(tf.squeeze(style_image))
plt.title("Style Image")
plt.axis("off")
plt.show()

# Load model
x = tf.keras.applications.vgg19.preprocess_input(content_image*255.0)
style_x = tf.keras.applications.vgg19.preprocess_input(style_image*255.0)

# Use TensorFlow's built-in NST model
hub_model = tf.keras.Sequential([
    tf.keras.layers.Lambda(lambda x: tf.image.resize(x, (256, 256))),
    tf.keras.layers.Rescaling(1./255)
])

import tensorflow_hub as hub
nst_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')

stylized_image = nst_model(content_image, style_image)[0]

plt.imshow(tf.squeeze(stylized_image))
plt.title("Stylized Output")
plt.axis("off")
plt.show()
```
